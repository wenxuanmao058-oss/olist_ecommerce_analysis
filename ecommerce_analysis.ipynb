```python
# 本地SQLite路径，请修改为你自己的路径
# 数据集下载：https://www.kaggle.com/datasets/olistbr/brazilian-ecommerce
import pandas as pd
import sqlite3
conn = sqlite3.connect(r"D:\项目\olist.sqlite")  # ← 改成你本地的路径
all_tables=pd.read_sql("SELECT name FROM sqlite_master WHERE type='table'", conn)
```

```python
orders = pd.read_sql("SELECT * FROM orders", conn)
order_items = pd.read_sql("SELECT * FROM order_items", conn)
order_payments = pd.read_sql("SELECT * FROM order_payments", conn)
customers = pd.read_sql("SELECT * FROM customers", conn)
products = pd.read_sql("SELECT * FROM products", conn)
order_reviews = pd.read_sql("SELECT * FROM order_reviews", conn)
```

```python
print("=== orders ===")
print(orders.shape)
print(orders.isnull().sum())
print(orders.duplicated().sum())
```

```python
print("=== order_items ===")
print(order_items.shape)
print(order_items.isnull().sum())
print(order_items.duplicated().sum())
```

```python
print("=== order_payments ===")
print(order_payments.shape)
print(order_payments.isnull().sum())
print(order_payments.duplicated().sum())
```

```python
print("=== customers ===")
print(customers.shape)
print(customers.isnull().sum())
print(customers.duplicated().sum())
```

```python
print("===products===")
print(products.shape)
print(products.isnull().sum())
print(products.duplicated().sum())
# 商品分类缺失的，标记为"未分类"
products['product_category_name'] = products['product_category_name'].fillna('未分类')

# 重量尺寸缺失的2行直接删除，数量很少不影响整体
products = products.dropna(subset=['product_weight_g'])

```

```python
print("===order_reviews===")
print(order_reviews.shape)
print(order_reviews.isnull().sum())
print(order_reviews.duplicated().sum())
```

```python
#日期转换，字符串转日期格式
date_columns = ['order_purchase_timestamp', 'order_approved_at', 
             'order_delivered_carrier_date', 'order_delivered_customer_date',
             'order_estimated_delivery_date']
for col in date_columns:
    orders[col] = pd.to_datetime(orders[col])
```

```python
# 只保留已完成的订单
orders_clean = orders[orders['order_status'] == 'delivered'].copy()
print(f"\n清洗后订单数：{len(orders_clean)}")
```

```python
# 删除重复值
orders_clean = orders_clean.drop_duplicates()
order_items = order_items.drop_duplicates()
```

```python
print(orders_clean.isnull().sum())
orders_clean = orders_clean.dropna(subset=['order_approved_at','order_delivered_carrier_date','order_delivered_customer_date'])
```

清洗完成合成宽表

```python
df = orders_clean.merge(order_items, on='order_id', how='inner')#订单主表+商品明细表，内连接，过滤商品明细为空的无效数据
df = df.merge(products, on='product_id', how='left')#左连接，关联商品信息表
df = df.merge(customers, on='customer_id', how='left')#关联用户信息，左连接

payments_agg = order_payments.groupby('order_id')['payment_value'].sum().reset_index()#按订单分组，金额求和，分组后索引变为普通索引
df = df.merge(payments_agg, on='order_id', how='left')

reviews_agg = order_reviews.groupby('order_id')['review_score'].mean().reset_index()#订单分组，评分求平均
df = df.merge(reviews_agg, on='order_id', how='left')

print(df.shape)
df.head()
```
